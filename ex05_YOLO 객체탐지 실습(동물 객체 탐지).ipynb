### 객체 탐지 평가지표
> AP(Average Precision)
> - P-R Curve(신뢰도 값에 따라 정밀도, 재현율 값을 기반으로 산점도를 그리고 꺾은 선으로 표현한 차트
> - 1ea의 클래스에 대해서 P-R Curve를 계단형 곡선으로 만들어주고, 그 면적(Area)을 구한 값
> -  0 ~ 1 사이의 값, 값이 높으면 높을수록 높은 정밀도와 재현율을 함께 유지한다는 뜻

> mAP(mean Average Precision) : 여러클래스에 대한 AP값의 평균

> mAP50 : IoU 임계값을 0.5로 두고 계산한 mAP 값(0.5 이상인 경우 정탐)  <<< 제일 많이 씀
